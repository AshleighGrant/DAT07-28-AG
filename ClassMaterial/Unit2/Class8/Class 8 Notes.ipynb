```python
df = pd.read_excel("superstore.xls")
```

```python
df
```

```python
sales_mean = df.groupby("Sub-Category")[["Sales"]].mean()
```

```python
sales_mean
```

```python
sales_mean.rename({"Sales": "SubCatAvg"}, axis=1, inplace=True)
```

```python
df = df.merge(sales_mean, left_on="Sub-Category", right_index=True)
```

```python
df["PercentDiff"] = (df["Sales"] - df["SubCatAvg"]) / df["SubCatAvg"]
```

```python
cust_avgs = df.groupby("Customer Name")["PercentDiff"].mean()*100
```

```python
cust_avgs.sort_values(ascending=False)
```

```python
profit_mean = df.groupby("Sub-Category")[["Profit"]].mean()
```

```python
profit_mean
```

```python
profit_mean.rename({"Profit": "SubCatProfitAvg"}, axis=1, inplace=True)
```

```python
df = df.merge(profit_mean, left_on="Sub-Category", right_index=True)
```

```python
df["PercentDiff"] = (df["Profit"] - df["SubCatProfitAvg"]) / df["SubCatProfitAvg"]
```

```python
cust__profit_avgs = df.groupby("Customer Name")["PercentDiff"].mean()
```

```python
cust__profit_avgs.sort_values(ascending=False)
```
