```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
```

```python
df = pd.read_csv("/Users/imac/DAT07-28-AG/ClassMaterial/Unit3/data/iowa_train2.csv")
```

```python
df.head()
```

```python
missing_cols_query = df.isnull().sum() > 0
missing_cols_num = df.loc[:, missing_cols_query].select_dtypes(include=np.number).columns.tolist()
missing_cols_cat = df.loc[:, missing_cols_query].select_dtypes(include=np.object).columns.tolist()
df[missing_cols_num] = df[missing_cols_num].fillna(0)
df[missing_cols_cat] = df[missing_cols_cat].fillna("None")
```

```python
# declare X and y
X = df.drop("SalePrice", axis=1)
y = df["SalePrice"]

# X = pd.get_dummies(X)
```

```python
X.head()
```

```python
gbm = GradientBoostingRegressor()
gbm.fit(X, y)
```

```python
gbm.score(X, y)
```

```python
# split your data into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1985)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train.head()
```

```python
# do all the work on the training set

gbm.fit(X_train, y_train)
```

```python
#always score on the test set
gbm.score(X_test, y_test)
```

```python
# validation set
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1985)
```

```python
gbm.fit(X_train, y_train)
```

```python
gbm.score(X_val, y_val)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(estimator=gbm, X=X_train, y=y_train, cv=10)
```
