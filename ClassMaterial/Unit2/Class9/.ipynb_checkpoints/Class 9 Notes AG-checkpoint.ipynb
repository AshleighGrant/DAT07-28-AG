```python
df[["Price"]].shift()
```

```python
df["Price"].mean(level=1)
```

```python
#for some operations, can specify which panel level you want to apply it to
df["Price"].mean(level=0)
```

```python
#if level=None, you can do the above. Won't work on all operations (e.g. shift)
?df.sum
```

```python
df.groupby(level=0)["Price"].shift().loc["MSFT"]
```

```python
df["1Day"] = df.groupby(level=0)["Price"].shift()
```

```python
df["2Day"] = df.groupby(level=0)["Price"].shift(2)
```
